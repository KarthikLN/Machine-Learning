```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("BMW sales data (2010-2024) (1).csv")
```

```python
df
```

```python
df["Sales_Classification"]=df["Sales_Classification"].map({"High":1,"Low":0})
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoder.fit_transform(df[["Model","Region","Color","Fuel_Type","Transmission"]]).toarray()
```

```python
df1=pd.DataFrame(encoder.fit_transform(df[["Model","Region","Color","Fuel_Type","Transmission"]]).toarray(),columns=encoder.get_feature_names_out())
```

```python
df2=pd.concat([df,df1],axis=1)
```

```python
df2
```

```python
df2.columns
```

```python
df2.drop(columns=["Model","Region","Color","Fuel_Type","Transmission"],inplace=True)
```

```python
df2.columns
```

```python
df2.dtypes
```

```python
X=df2.drop("Sales_Classification",axis=1)
y=df2["Sales_Classification"]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=34)
```

```python
rf_cls=RandomForestClassifier()
```

```python
param_grid={
    "n_estimators":[50,100,200,300],
    "max_depth":[10,20,30],
    "criterion":["gini","entropy","log_loss"] #(36 grids * 5 fold).
}

gridSearch=GridSearchCV(estimator=rf_cls,param_grid=param_grid,cv=5,n_jobs=1,verbose=2)

gridSearch.fit(X_train,y_train)
```

```python
gridSearch.best_params_
```

```python
RFC=RandomForestClassifier(criterion="gini",max_depth=10,n_estimators=50,random_state=42)
```

```python
RFC.fit(X_train,y_train)
```

```python
RFC.predict(X_test)
y_pred=RFC.predict(X_test)
print(classification_report(y_test,y_pred))
```
